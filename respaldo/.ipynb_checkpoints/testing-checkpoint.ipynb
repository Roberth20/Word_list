```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
resul = [[pd.to_datetime(dt.date.today()), nota[0], nota[1], nota[2], nota[3], nota[4],final]]
```

```python
df = pd.DataFrame(resul, columns = ["Fecha","Level 1", "Level 2", "Level 3", "Level 4", "Level 5", "Final"])
```
